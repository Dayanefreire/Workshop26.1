```python
import numpy as np
temperaturas = np.random.uniform(20, 35, 7).round(1)
dias = ['segunda', 'terça', 'quarta', 'quinta', 'sexta', 'sábado', 'domingo']
print(temperaturas)
print(dias)
```

```python
import pandas as pd
df = pd.DataFrame({
    'dia': dias,
    'temperatura': temperaturas
})

print(df)

media = df['temperatura'].mean().round(1)
maxima = df['temperatura'].max()
minima = df['temperatura'].min()

print(f"\nMédia: {media}ºC")
print(f"Máxima: {maxima}ºC")
print(f"Mínima: {minima}ºC")
```

```python
import  matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
plt.plot(dias, temperaturas, marker='o', color='tomato')
plt.title('Temperatura ao longo da semana')
plt.xlabel('Dia')
plt.ylabel('Temperatura (°C)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.bar(dias, temperaturas, color='steelblue')
plt.title('Temperatura por dia')
plt.xlabel('Dia')
plt.ylabel('Temperatura (°C)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.fill_between(range(7), temperaturas, alpha=0.4, color='green')
plt.plot(range(7), temperaturas, color='green', marker='o')
plt.title('Variação de Temperatura')
plt.xlabel('Dia')
plt.ylabel('Temperatura (°C)')
plt.xticks(range(7), dias, rotation=45)
plt.tight_layout()
plt.show()
```
